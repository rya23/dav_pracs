```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url,index_col=0,header=0,parse_dates=True)
df.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
plt.figure(figsize=(17,6))
plt.plot(df['Passengers'])
plt.title("Data")
plt.show()
```

```python
df['SMA_12'] = df['Passengers'].rolling(window=12).mean()

# Plot with SMA
plt.plot(df['Passengers'], label='Original')
plt.plot(df['SMA_12'], label='12-Month SMA', color='red')
```

```python
result = seasonal_decompose(df,period=30,model='additive')
result.plot()
plt.show()
```

```python
result = seasonal_decompose(df,period=30,model='additive')
plt.figure(figsize=(17,6))
plt.plot(result.trend)
plt.title("Trends")
plt.show()
```

```python
plt.figure(figsize=(17,6))
plt.plot(result.seasonal)
plt.title("Seasonality")
plt.show()
```

```python
plt.figure(figsize=(17,6))
plt.plot(result.resid)
plt.title("Residuals")
plt.show()
```
